```python
def detect(img):
    out = np.zeros(img.shape).astype(np.int32)
    for i in range(img.shape[0]-1):
        for j in range(img.shape[1]):
            out[i,j] += abs(int(img[i,j]) - int(img[i+1,j]) ) 
            
    plt.imshow(out)
    return plt.show()
```
