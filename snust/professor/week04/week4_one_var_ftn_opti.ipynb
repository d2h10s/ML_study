```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def numerical_D(f, x, h=1e-6):
    return (f(x+h)-f(x))/h
```

```python
def numerical_D2(f, x, h=1e-6):
    return (f(x+h)-f(x-h))/(2*h)
```

```python
f1 = lambda x : x**2
print(numerical_D(f1, 1))
print(numerical_D2(f1, 1))
```

```python
class descent:
    def __init__(self, lr, h=1e-6):
        self.lr = lr
        self.h = h
    
    def update(self, f, x):
        df =  numerical_D2(f, x, self.h)
        return x - self.lr*df
        
    def __str__(self):
        return '강하법'
```

```python
opti1 = descent(0.01)
print(opti1)
```

```python
class one_variable_ftn:
    
    def __init__(self, ftn, opti):
        self.ftn = ftn
        self.opti = opti
        
    def run(self, x0, iteration):
        self.xs = [x0]
        for j in range(iteration):
            self.xs.append(self.opti.update(self.ftn, self.xs[-1]))
        print(""" %s에서 시작해서 %s번 %s을 적용한 결과는 %.3f입니다.
        """%(x0, iteration, self.opti, self.xs[-1]))
        pass

    def visualize(self, xi, xe):
        x = np.linspace(xi,xe,int((xe-xi)*100))
        y = [self.ftn(xx) for xx in x]
        plt.plot(x, y, 'k-')
        ys = [self.ftn(xx) for xx in self.xs] 
        plt.plot(self.xs, ys, 'r*')
        return plt.show()

    def history(self):
        for j,x in enumerate(self.xs):
            print("%s번 %s를 적용한 결과는 %.3f" %(j, self.opti,  x))
        pass
```

```python
ex1 = one_variable_ftn(f1, opti1)
```

```python
ex1.run(2, 100)
```

```python
ex1.visualize(-1, 2.5)
```

```python
ex1.history()
```

```python
opti2 = descent(0.1)
ex2 = one_variable_ftn(f1, opti2)
ex2.run(2, 40)
```

```python
opti3 = descent(2)
ex3 = one_variable_ftn(f1, opti3)
ex3.run(2, 5)
ex3.history()
```

```python
f2 = lambda x : x**4
ex4 = one_variable_ftn(f2,opti1)
ex4.run(2,200)
ex4.visualize(-1,3)
```
