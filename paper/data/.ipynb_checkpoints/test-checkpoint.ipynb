```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
angle = pd.read_csv('angle.csv')
loss = pd.read_csv('losses.csv')
reward = pd.read_csv('rewards.csv')
```

```python
angle.head()
```
